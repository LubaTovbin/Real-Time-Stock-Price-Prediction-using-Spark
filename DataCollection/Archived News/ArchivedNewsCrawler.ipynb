```python
#Importing the necessary libraries
import concurrent.futures
import calendar
import requests
from datetime import datetime
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
from newspaper import Article
import newspaper
from langdetect import detect
```

```python
#URL to crawl from
URL = "https://www.google.com/search?q={query}&tbas=0&biw=1536&bih=754&source=lnt&tbs=cdr%3A1%2Ccd_min%3A{month}%2F{day}%2F{year}%2Ccd_max%3A{month}%2F{day}%2F{year}&tbm=nws"
headers={"User-Agent":"Mozilla/5.0 (X11; Linux x86_64; rv:60.0) Gecko/20100101 Firefox/60.0","Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8"}
```

```python
def downloadContent(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        print("Downloading from..",url)
        if(detect(article.text)!='en'):
            return ""
        if(len(article.text) > 350 and len(article.text)<6000):
            return article.text
        return ""
    except Exception:
        print("in exc")
        return ""
```

```python
def downloadNews(urls,date,query):
    #Creating multiple threads to crawl simultaneously
    newsData =pd.DataFrame()
    with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
        jobs = {executor.submit(downloadContent, url):url for url in urls}
        for future in concurrent.futures.as_completed(jobs):
            try:
                data = future.result()
                if data!="":
                    df =pd.DataFrame({'URL':jobs[future],'News':data,'Type':query,'Date':date},index=[0])
                    newsData = pd.concat([df, newsData], ignore_index=True, sort = False)
            except Exception as exc:
                print('generated an exception: %s' % (exc))   
    return newsData
```

```python
def load_data(**params):
    day = int(params.get('day'))
    year = int(params.get('year'))
    month = int(params.get('month'))
    query = params.get('query')
    session=requests.session()
    response=session.get(URL.format(**params),headers=headers)
    soup=BeautifulSoup(response.text,"html.parser")
    divs = soup.findAll("div",{"class":"dbsr"})
    newsUrls = []
    for div in divs:
        url = div.find("a")["href"]
        newsUrls.append(url)
    date = datetime(year,month,day)
    return downloadNews(newsUrls,date,query)
```

```python
#Creating multiple threads to crawl the URL's simultaneously
with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
    year = 2018
    for month in range(1,13):
        num_days = calendar.monthrange(year, month)[1]
        jobs = [executor.submit(load_data, query='microsoft',year=year,month=month,day=day) for day in range(1,num_days+1)]
        for future in concurrent.futures.as_completed(jobs):
            try:
                newsContent = future.result()
                with open('archived.csv', 'a+', encoding="utf-8-sig", newline='') as f:
                    newsContent.to_csv(f, header=False,index=False)
            except Exception as exc:
                print('generated an exception: %s' % (exc))
                
```
